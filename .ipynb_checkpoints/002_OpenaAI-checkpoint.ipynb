## Esempio di utilizzo della OpenAI ChatCompletion API

Sono definiti 3 ruoli:
- System (serve per settare il comportamento dell'assistant)
- User (contiene la richiesta fatta all'assistant)
- Assistant (Contiene le precedenti risposte)

Per utilizzare l'API da Langchain devo quindi importare lo schema

```python
from langchain.schema import (
    SystemMessage,
    AIMessage,
    HumanMessage)

messages =[
    SystemMessage(content="you are a fhisician and respond only in german"),
    HumanMessage(content="Spiegami la meccanica quantistica")
]

output = llm.invoke(messages)
print(output.content)
```
